Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

soln:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter plot of x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

soln:
    

The subplot() function is used to create multiple plots in a single figure. It allows us to create multiple axes or subplots in the same figure and arrange them in a grid-like structure.

Here's an example of how to create four line plots using the subplot() function:

```python


import matplotlib.pyplot as plt
import numpy as np

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add overall title and axis labels
fig.suptitle('Four Line Plots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

In this code, we first create four arrays (x1, y1, x2, y2, x3, y3, x4, y4) to hold the data for each line plot. Then, we create a figure with 2 rows and 2 columns of subplots using the subplots() function. We plot each line on a different subplot using the plot() function and set the title for each subplot using the set_title() function. Finally, we add an overall title and axis labels to the figure using the suptitle(), xlabel(), and ylabel() functions.

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

soln:

A bar plot is a way of visualizing data using rectangular bars. The length of each bar is proportional to the value of the item being represented. Bar plots are used to compare different categories of data or to show the relationship between a categorical variable and a numerical variable.

To plot a bar plot and a horizontal bar plot using the given data, we can use the bar() and barh() functions of the Pyplot module of Matplotlib, respectively.

Here's the code to plot a bar plot and a horizontal bar plot for the given data:

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title("Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit ($ billions)")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Profit ($ billions)")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

soln:

A box plot is a statistical chart used to display the distribution of a set of data. It shows the median, quartiles, and outliers of the data. It is used to identify the shape and spread of the data and to detect outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
fig, ax = plt.subplots()
ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_title('Box Plot')
ax.set_ylabel('Values')
plt.show()
```

we have used the boxplot() function to create a box plot. We have created a list data containing the two arrays box1 and box2. We have then created a subplots object using subplots() function and passed it to the ax variable. We have then plotted the boxplot using ax.boxplot(data). Finally, we have set the title, x-label, y-label, and tick labels using the set_title(), set_xlabel(), set_ylabel(), and set_xticklabels() functions.
